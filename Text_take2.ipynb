```python
# IMPORT DEPENDENCIES
# Import dependencies
import numpy as np
import pandas as pd
import os
```

```python
# READ IN FILE
df = pd.read_csv('Tables/free_text_df.csv')
df.head(10)
```

```python
df.info()
```

```python
# Extract WhyStopped column and drop null values
df_text = pd.DataFrame(df['WhyStopped'])
df_text = df_text.dropna()
df_text
```

```python
df_text.info()
```

```python
df_text['word_count'] = df_text['WhyStopped'].apply(lambda x: len(str(x).split(" ")))
df_text.head(10)
```

```python
# AVERAGE WORD COUNT
df_text['word_count'].mean()
```

```python
# IMPORT STOPWORD LIBRARY
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
df_text['stop_words'] = df_text['WhyStopped'].apply(
    lambda x: len([x for x in x.split() if x in stop]))
df_text.head(10)
```

```python
# AVERAGE STOPWORDS
df_text['stop_words'].mean()
```

```python
# LOWERCASE RESPONSES
df_text['WhyStopped'] = df_text['WhyStopped'].apply(
    lambda x: " ".join(x.lower() for x in x.split()))
df_text.head(10)
```

```python
# REMOVING PUNCTUATION
df_text['WhyStopped'] = df_text['WhyStopped'].str.replace('[^\w\s]','')
df_text.head(10)
```

```python
# REMOVE STPWORDS
df_text['filtered_responses'] = df_text['WhyStopped'].apply(
    lambda x: " ".join(x for x in x.split() if x not in stop))
df_text.head(10)
```

```python
# WORD FREQUENCY
freq = pd.Series(' '.join(df_text['filtered_responses']).split()).value_counts()[:30]
freq
```

```python
rate_freq = df_text.loc[df_text['WhyStopped'].str.contains("rate", case=False)]
rate_freq.sample(20)
```
